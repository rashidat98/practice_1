```python
# Import the neccessary python library
import pandas as pd
import numpy as np
```

```python
# input the given data into a dataframe named df
data = {
'Name': ['Alice', 'Bob', 'Charlie', None, 'Eve'],
'Age': [24, None, 22, 32, 29],
'City': ['New York', 'Los Angeles', None, 'Houston', 'Phoenix'],
'Salary': [70000, 80000, None, 120000, None]
}

```

```python
# Display the dataframe
df = pd.DataFrame(data)
df
```

```python
# count the missing values present in each column in the df dataframe using .isna().sum()
df.isna().sum()
```

```python
# create a new column called salary_filled and fill the missing values with the average salary
average_salary = df["Salary"].mean()
df["salary_filled"] = df["Salary"].fillna(average_salary)
df
```
